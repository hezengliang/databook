## 地图显示模块
ipyleaflet是基于leaflet的地图显示模块，提供了jupyterlab支持。
- Tutorial, https://my.oschina.net/u/2306127/blog/1859080
- ipyleaflet文档，https://ipyleaflet.readthedocs.io/en/latest/installation.html#jupyterlab-extension
- GeoJson绘制，https://github.com/jupyterlab/jupyter-renderers/tree/master/packages/geojson-extension
- 流体地图可视化，https://github.com/benbovy/jupyterlab_velocity

安装到jupyterlab，执行：

conda install -c conda-forge nodejs=8.10.0 -y

conda install -c conda-forge ipyleaflet -y

jupyter labextension install jupyter-leaflet
